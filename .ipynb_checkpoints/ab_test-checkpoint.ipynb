```python
# data cleaning, change null with 0
merge_df['spent'] = merge_df['spent'].fillna(0)

df_peruser = merge_df.groupby('uid_x').agg({'spent': 'mean', 'group': 'first'})
df_peruser['sum_per_user'] = merge_df.groupby('uid_x')['spent'].sum()


print(df_peruser)
```

```python

# average spent per user group A, control
average_spent_A = df_peruser.loc[df_peruser['group'] == 'A', 'sum_per_user'].mean().round(2)
print('Average spent Group A: ' + str(average_spent_A))

average_spent_B = df_peruser.loc[df_peruser['group'] == 'B', 'sum_per_user'].mean().round(2)
print('Average spent Group B: ' + str(average_spent_B))
```

```python
# standard deviation
std_a = df_peruser.loc[df_peruser['group'] == 'A', 'sum_per_user'].std().round(2)
std_b = df_peruser.loc[df_peruser['group'] == 'B', 'sum_per_user'].std().round(2)

print("Standard deviation group A: " + str(std_a))
print("Standard deviation group B: " + str(std_b))

# n sample A, B
n_A = df_peruser.loc[df_peruser['group'] == 'A','group'].count()
n_B = df_peruser.loc[df_peruser['group'] == 'B','group'].count()
print('N sample A:', n_A)
print('N sample B:', n_B)

# sample A, B average spent 
sample_A_avg = df_peruser.loc[df_peruser['group'] == 'A','spent']
sample_B_avg =  df_peruser.loc[df_peruser['group'] == 'B','spent']

# sample A,B sum spent

sample_A_sum = df_peruser.loc[df_peruser['group'] == 'A','sum_per_user']

sample_B_sum = df_peruser.loc[df_peruser['group'] == 'B', 'sum_per_user']
```

```python
# specify the desired confidence level (e.g. 95%)
confidence_level = 0.95

# calculate the critical value of t
degrees_of_freedom = n_A
t_critical = t.ppf(confidence_level, degrees_of_freedom)

# calculate the margin of error
margin_of_error = t_critical * (std_a / np.sqrt(n_A))

# calculate the confidence interval
confidence_interval = (average_spent_A - margin_of_error, average_spent_A + margin_of_error)

# print the results
print(f"Sample mean: {average_spent_A:.2f}")
print(f"Sample standard deviation: {std_a:.2f}")
print(f"Critical value of t: {t_critical:.2f}")
print(f"Margin of error: {margin_of_error:.2f}")
print(f"Confidence interval: {confidence_interval}")
```

```python
# a two-sample t-test
t_statistic, p_value = stats.ttest_ind(sample_A_sum, sample_B_sum)
print('t-statistic', t_statistic)
```
